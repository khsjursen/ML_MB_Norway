```python
def aggregate_ba_by_periods(df):
    # Filter data for the years within 1980-2019
    period_filter = df['Year'].between(1980, 2019)
    df_filtered = df[period_filter]

    # Define the function to determine the period
    def determine_period(year):
        if 1980 <= year <= 1989:
            return '1980-1989'
        elif 1990 <= year <= 1999:
            return '1990-1999'
        elif 2000 <= year <= 2009:
            return '2000-2009'
        elif 2010 <= year <= 2019:
            return '2010-2019'
        else:
            return None

    df_filtered['Period'] = df_filtered['Year'].apply(determine_period)

    # Ensure counts of years per period and RGIID are as expected (10 years for each period)
    expected_counts = {'1980-1989': 10, '1990-1999': 10, '2000-2009': 10, '2010-2019': 10}
    period_counts = df_filtered.groupby(['RGIID', 'Period'])['Year'].nunique()

    # Filter out periods where year data is incomplete
    complete_periods = period_counts[period_counts == period_counts.index.map(lambda x: expected_counts[x[1]])].index

    # Filter df based on complete periods
    df_complete_periods = df_filtered[
        df_filtered.set_index(['RGIID', 'Period']).index.isin(complete_periods)
    ]

    # Group by RGIID, BREID, and Period and aggregate Ba
    grouped = df_complete_periods.groupby(['RGIID', 'BREID', 'Period'])['Ba','Area'].mean().reset_index()
    grouped.rename(columns={'Ba': 'Ba_Glaciological'}, inplace=True)
    
    return grouped

aggregated_obs_df = aggregate_ba_by_periods(df_obs)

aggregated_obs_df
```

```python
# Get data for common time period and merge with observations

test_glaciers = ['00434', '00449', '01013', '01026', '01258', '01286', '01484', '01598', '02384', '02650', '02916', '02918', '02920', '02969']
train_glaciers = ['00287','00295','00868','00966','00987','01126','01186','01217','01657','01779','02017','02436','02458','02643','02666','02962','02963','02966']
all_glaciers = test_glaciers + train_glaciers

# Set the start and end years for the filter
start_year = 1962
end_year = 2021

gl = 'test'

# Function to filter DataFrame based on RGIID ending and year range
def filter_df_by_rgiid_and_year(df, test_glaciers, start_year, end_year):
    return df[df['RGIID'].str.endswith(tuple(test_glaciers)) & (df['Year'] >= start_year) & (df['Year'] <= end_year)]

# Filter the DataFrames
#df_obs_nonan = df_obs.dropna(subset=['RGIID'])

if gl == 'test':
    df_pred_filtered = filter_df_by_rgiid_and_year(df_pred, test_glaciers, start_year, end_year)
elif gl == 'all':
    df_pred_filtered = filter_df_by_rgiid_and_year(df_pred, all_glaciers, start_year, end_year)

# Function to rename columns based on model name
def rename_columns(df, model_name):
    return df.rename(columns={
        'Ba': f'Ba_{model_name}',
        'Bs': f'Bs_{model_name}',
        'Bw': f'Bw_{model_name}'
    })

# Renaming columns of the filtered DataFrames
df_pred_filtered = rename_columns(df_pred_filtered, 'ML')

df_pred_filtered
```

```python
# Extract relevant columns
ba_columns = ['RGIID', 'Year'] + [col for col in df_pred_filtered.columns if col.startswith('Ba_')]
ba_df = df_pred_filtered[ba_columns]

# Define the 10-year bins
bins = list(range(1960, 2021, 10))  # From 1980 to 2020, step by 10
labels = [f"{start}-{start+9}" for start in bins[:-1]]

# Create a 'Period' column by cutting the 'Year' into bins
ba_df['Period'] = pd.cut(ba_df['Year'], bins=bins, labels=labels, right=False)

# Group by 'RGIID' and 'Period' and then calculate the mean for each
aggregated_df = ba_df.groupby(['RGIID', 'Period']).mean().reset_index()

# Drop the 'Year' column if it's still present
if 'Year' in aggregated_df.columns:
    aggregated_df = aggregated_df.drop(columns=['Year'])

aggregated_df
```

```python
# Glaciological observations
filepath_obs = 'C:/Users/kasj/ML_MB_Norway/Data/NVE_gw_mb/massbalance_gw_Norway.csv'
df_obs = pd.read_csv(filepath_obs, sep=';')

# Filter glaciological obs on RGIID
df_obs = df_obs[(df_obs['RGIID'].notna())]
df_obs = df_obs[(df_obs['RGIID'].str.endswith(tuple(test_glaciers)))]
```

```python
def aggregate_ba_by_periods(df):
    # Filter data for the years within 1980-2019
    period_filter = df['Year'].between(1960, 2021)
    df_filtered = df[period_filter]

    # Define the function to determine the period
    def determine_period(year):
        if 1960 <= year <= 1969:
            return '1960-1969'
        elif 1970 <= year <= 1979:
            return '1970-1979'
        elif 1980 <= year <= 1989:
            return '1980-1989'
        elif 1990 <= year <= 1999:
            return '1990-1999'
        elif 2000 <= year <= 2009:
            return '2000-2009'
        elif 2010 <= year <= 2019:
            return '2010-2019'
        else:
            return None

    df_filtered['Period'] = df_filtered['Year'].apply(determine_period)

    # Ensure counts of years per period and RGIID are as expected (10 years for each period)
    n_yrs = 10
    expected_counts = {'1960-1969': n_yrs, '1970-1979': n_yrs, '1980-1989': n_yrs, '1990-1999': n_yrs, '2000-2009': n_yrs, '2010-2019': n_yrs}
    period_counts = df_filtered.groupby(['RGIID', 'Period'])['Year'].nunique()

    # Filter out periods where year data is incomplete
    complete_periods = period_counts[period_counts == period_counts.index.map(lambda x: expected_counts[x[1]])].index

    # Filter df based on complete periods
    df_complete_periods = df_filtered[
        df_filtered.set_index(['RGIID', 'Period']).index.isin(complete_periods)
    ]

    # Group by RGIID, BREID, and Period and aggregate Ba
    grouped = df_complete_periods.groupby(['RGIID', 'BREID', 'Period'])['Ba','Area'].mean().reset_index()
    grouped.rename(columns={'Ba': 'Ba_Glaciological'}, inplace=True)
    
    return grouped

aggregated_obs_df = aggregate_ba_by_periods(df_obs)

aggregated_obs_df
```

```python
# Merge with dataframe with model results
# Merge the dataframes on RGIID and Period
# aggregated_obs_df_ = aggregated_obs_df.drop(columns=['BREID','Area'])
aggregated_df_all_obs = pd.merge(aggregated_df, aggregated_obs_df_, how='left', on=['RGIID', 'Period'])

aggregated_df_all_obs
```

```python
from sklearn.metrics import root_mean_squared_error, mean_squared_error, mean_absolute_error, r2_score

# Filter rows where Ba_Glaciological is not NaN
filtered_df = aggregated_df_all_obs.dropna(subset=['Ba_Glaciological'])

y_true = filtered_df['Ba_Glaciological']
y_pred = filtered_df['Ba_ML']

rmse_value = root_mean_squared_error(y_true, y_pred)
mae_value = mean_absolute_error(y_true, y_pred)
r2_value = r2_score(y_true, y_pred)
bias_value = np.mean(y_pred - y_true)

print(f'RMSE: {rmse_value}')
print(f'MAE: {mae_value}')
print(f'R²: {r2_value}')
print(f'Bias: {bias_value}')
```

```python
# Import libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.lines as mlines
import xarray as xr
from pandas.tseries.offsets import MonthBegin
```

```python
# Get dataframe of geodetic observations
filepath_obs = 'C:/Users/kasj/ML_MB_Norway/Data/NVE_gw_mb/Bgeod_Norway.csv'

df_obs = pd.read_csv(filepath_obs, sep=';')
df_obs
```

```python
# Get monthly glacier-wide mass balances for all models

# Read dataframe of ML predictions
filepath_ML = 'C:/Users/kasj/ML_MB_Norway/Data/glacier_predictions/with_predictions_monthly_2024-10-24_10-30-28/ML_monthly_mb_glacierwide.nc'
#filepath_ML = 'C:/Users/kasj/ML_MB_Norway/Data/glacier_predictions/with_predictions_monthly_2024-10-25_09-45-30/ML_monthly_mb_glacierwide.nc'

with xr.open_dataset(filepath_ML) as ds_ML_out:
    ds_ML = ds_ML_out

# OGGM
filepath_OGGM = 'C:/Users/kasj/ML_MB_Norway/Data/data_model_comparison/OGGM_monthly_mb_glacierwide.nc'
with xr.open_dataset(filepath_OGGM) as ds_OGGM_out:
    ds_OGGM = ds_OGGM_out

# GloGEM
filepath_GLOGEM = 'C:/Users/kasj/ML_MB_Norway/Data/data_model_comparison/GloGEM_monthly_mb_glacierwide.nc'
with xr.open_dataset(filepath_GLOGEM) as ds_GLOGEM_out:
    ds_GLOGEM = ds_GLOGEM_out

# PyGEM
filepath_PYGEM = 'C:/Users/kasj/ML_MB_Norway/Data/data_model_comparison/PyGEM_monthly_mb_glacierwide.nc'
with xr.open_dataset(filepath_PYGEM) as ds_PYGEM_out:
    ds_PYGEM = ds_PYGEM_out

```

```python
# Aggregate montlhy mass balance between and including start and end months for given RGIID and model
def compute_aggregated_metric(ds, RGIID_last_four_digits, from_date, to_date):

    subset = ds.sel(ID=RGIID_last_four_digits, time=slice(from_date, to_date))
    
    mb_sum = subset.mb_monthly.sum().item()
    
    years = (to_date.year - from_date.year)

    # Divide by whole year
    metric = mb_sum / years
    
    return metric

# Iterate through each row, compute metrics, and store results
datasets = {'MBM': ds_ML, 'OGGM': ds_OGGM, 'PYGEM': ds_PYGEM, 'GLOGEM': ds_GLOGEM}

# Run thorough dates and predict mass balance rate for each model
df_results = df_obs.copy()

model_columns = ['MBM', 'OGGM', 'PYGEM', 'GLOGEM']
for col in model_columns:
    df_results[col] = np.nan

for index, row in df_results.iterrows():
    RGIID_last_four_digits = row['RGIID'][-4:]
    num_models = row['num_models']
    
    if num_models == 0 or num_models == 1:
        model_names = ['MBM']
    elif num_models == 4:
        model_names = ['MBM', 'OGGM', 'PYGEM', 'GLOGEM']
    
    from_date = pd.to_datetime(row['from_date_rounded'])
    to_date = pd.to_datetime(row['to_date_rounded'])
    
    for model_name in model_names:
        ds = datasets[model_name]
        metric = compute_aggregated_metric(ds, RGIID_last_four_digits, from_date, to_date)
        column_name = f'{model_name}'
        df_results.at[index, column_name] = metric
```
